```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(0,100,100)
y= np.cos(x)
z= np.sin(x)
```

```python
fig,axes=plt.subplots(ncols=2);
axes[0].plot(x,y);
axes[0].set_xlabel("x values");
axes[0].set_ylabel(" Sin / Cos");
plt.tight_layout();
plt.xticks([], []);
plt.yticks([], []);
```

```python
fig_save=axes[0].get_figure()
fig_save.savefig("fig.pdf",bbox_inches="tight",dpi=100)
```
