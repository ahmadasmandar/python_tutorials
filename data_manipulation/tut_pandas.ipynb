```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb 
```

```python
file=pd.read_csv("heart.csv")
df=pd.DataFrame(file)

%matplotlib inline
```

```python
chest_pain=df.groupby("cp").median().reset_index()
chest_pain
```

```python
chest_pain.plot.bar(x='cp');
```

```python
fig,ax=plt.subplots()
ax.bar(chest_pain['cp'],chest_pain['age'], label='age',edgecolor='k',color='red')
ax.set_xlabel("cp")
ax.legend();
```

```python
ax=sb.barplot('cp','age',errcolor='k',capsize=0.1,data=df)
```

```python
df.plot.scatter('age','trestbps');
```

```python
fig,ax=plt.subplots()
ax.scatter(df['age'],df['trestbps'], marker='^',edgecolor='k',c=df['age'],s=50,alpha=0.5)
ax.set_xlabel("cp")
ax.legend();
```

```python
sb.scatterplot('age','trestbps',hue='age',data=df,s=30,edgecolor='none');
```

```python
ages=df.groupby('age').median().reset_index()
```

```python
ages.plot.line('age', ['chol','trestbps']);
```

```python
fig,ax=plt.subplots()
ax.plot(ages['age'],ages['trestbps'],ls=':',lw=2,color='g')
ax.set_xlabel("age")
ax.legend();
```

```python
sb.lineplot('age','trestbps',data=ages)
```
