```
def preprocess_image(filename):
    img=load_img(filename,target_size=(299,299))
    display(img)
    img_array_expanded=np.expand_dims(img_to_array(img),axis=0)
    prepro_img=preprocess_input(img_array_expanded)
    return prepro_img
```
